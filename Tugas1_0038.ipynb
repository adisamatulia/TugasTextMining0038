```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
factory = StopWordRemoverFactory()
stopword = factory.create_stop_word_remover()
# Kalimat
#"a favorite place to have a delicious hot crispy chicken. At the moment, the size of the rice is becoming small than before. Limited chairs to be sited if you want to dine in." >> oiginal word
kalimat = "tempat favorit untuk makan ayam krispi panas yang enak. Saat ini, ukuran beras menjadi lebih kecil dari sebelumnya. Kursi terbatas untuk ditempatkan jika Anda ingin makan di tempat."
stop = stopword.remove(kalimat)
print(stop)
```

```python
#stemming menggunakan Sastrawi
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory# create stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()# stemming process
sentence = 'tempat favorit untuk makan ayam krispi panas yang enak. Saat ini, ukuran beras menjadi lebih kecil dari sebelumnya. Kursi terbatas untuk ditempatkan jika Anda ingin makan di tempat.'
output   = stemmer.stem(sentence)
print(output)
```
